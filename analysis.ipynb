```python
import pandas as pd
```

```python
file_path = "data/dfp_cia_aberta_2021/dfp_cia_aberta_DRE_con_2021.csv"
```

```python
df = pd.read_csv(file_path, delimiter=";", encoding="latin1")

```

```python
df.ESCALA_MOEDA.unique()
```

```python
df['VL_REAL'] = df.apply(lambda x: x.VL_CONTA if x.ESCALA_MOEDA == "UNIDADE" else x.VL_CONTA * 1000, axis=1)
```

```python
df[(df.CD_CONTA=='3.01')&(df.ORDEM_EXERC=='ÚLTIMO')].sort_values(by='VL_REAL', ascending=False).head(10)
```

```python
df.VL_REAL = df.VL_REAL.astype(int)
```

```python
denom_cia = "JBS S.A."
report_columns = ['DENOM_CIA', 'CD_CONTA', 'DS_CONTA', 'VL_REAL']
# df_filters = (df.CD_CONTA.apply(len) == 4) & (df.ORDEM_EXERC == 'ÚLTIMO') & (df.DENOM_CIA == denom_cia)
df_filters = (df.CD_CONTA.apply(len) == 4) & (df.ORDEM_EXERC == 'ÚLTIMO')
df[df_filters][report_columns]
test_df = df[df_filters][report_columns]
```

```python
test_df
```

```python
test_df.loc[test_df.CD_CONTA=='3.01'].sort_values(by='DENOM_CIA')[['DENOM_CIA','VL_REAL']]
```

```python
receita = test_df.loc[test_df.CD_CONTA=='3.03'].sort_values(by='DENOM_CIA')[['DENOM_CIA','VL_REAL']]
receita.rename(columns={'VL_REAL':'RECEITA'}, inplace=True)
receita.reset_index(drop=True, inplace=True)
lucro = test_df.loc[test_df.CD_CONTA=='3.11'].sort_values(by='DENOM_CIA')[['DENOM_CIA','VL_REAL']]
lucro.rename(columns={'VL_REAL':'LUCRO'}, inplace=True)
lucro.reset_index(drop=True, inplace=True)
```

```python
agg_dfp = receita.merge(lucro, how='inner', on='DENOM_CIA')
```

```python
agg_dfp['PE'] = agg_dfp.RECEITA / agg_dfp.LUCRO
```

```python
agg_dfp[(agg_dfp.PE>1)].sort_values(by='PE').head(20)
```
